```python
import pandas as pd
import numpy as np

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import joblib
import warnings 
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv("finance.csv")
dataset.head()
```

```python
x = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
x.head()
```

```python
y.head()
```

```python
numeric_feature = x.select_dtypes(exclude=["object"]).columns
catag_feature = x.select_dtypes(include=["object"]).columns
```

```python
numeric_feature
```

```python
catag_feature
```

```python
numeric_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

```python
catag_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(drop="first", handle_unknown="ignore", sparse_output=False))
])
```

```python
preprocess_pipeline = ColumnTransformer([
    ("numeric", numeric_pipeline, numeric_feature),
    ("catag", catag_pipeline, catag_feature)
])
```

```python
clean = preprocess_pipeline.fit(x)
```

```python
Clean_data = pd.DataFrame(clean.transform(x))
```

```python
Clean_data.head()
```

```python
joblib.dump(clean, "knn_preprocess_pipeline.pkl")
```

```python
joblib.load("knn_preprocess_pipeline.pkl")
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    Clean_data, y, test_size=0.2, random_state=42
)
```

```python
model = KNeighborsClassifier(
    n_neighbors=7,
    weights='distance',
    metric='minkowski',
    p=2
)
```

```python
model.fit(x_train, y_train)
```

```python
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
print("Training Accuracy:", accuracy_score(y_train, y_train_pred))
print("Testing Accuracy:", accuracy_score(y_test, y_test_pred))
```

```python
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_test_pred))
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_test_pred))
```

```python
joblib.dump(model, "knn.pkl")
```
